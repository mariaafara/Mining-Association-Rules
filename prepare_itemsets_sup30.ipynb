```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

# This notebook shows the scripts and analysis at the same time for itemsets of min support of 0.3

created a dataframe containing each itemset as a list and its sup 

```python
df=pd.DataFrame()
x=0
with open("results/decoded_itemsets_minsup30.txt") as f:
    for line in f:
        x=x+1
        arr=line.split("#")
        sup=int(arr[1].split(" ")[1])
        l=[]
        for i in arr[0].strip().replace(", ",",").split(" "):
            i=i.split(",")
            l.append((i[0][2:len(i[0])-1],i[1][1:len(i[1])-2]))
        df=df.append({"itemset":l, 'sup':sup},ignore_index=True)
        
```

```python
df.head()
```

created a new coloumn containing the length of each length (size of itemset)

```python
df["length"]=df.itemset.map(len)
```

sorted the rows by the itemset size then the sup (i.e if 2 itemsets have the same size they are sorted according to their supports)

```python
df=df.sort_values(by=["length","sup"],ascending=False)
```

```python
df=df.reset_index(drop=True)
```

```python
pd.set_option('max_colwidth', 800)
```

```python
df.head(10)
```

separeted each item in the itemset and created for each a new coloum to ease the analysing of the itemsets

```python
df2=pd.DataFrame()
for _,val in df.iterrows():
    dic={"itemset":val["itemset"],
         "sup":val["sup"],
         "length":val["length"]}
    for i in val["itemset"]:
        dic[i[0]]=i[1]
    df2=df2.append(dic,ignore_index=True)
```

```python
df2.head()
```

Extracting all the unique values from the most frequent attributes in all the itemsets

```python
cols=[]
for i in df2.columns:
    if i not in ["itemset","length","sup"]:
        cols.append(i)
        print(i,df2[i].unique())
cols.append("itemset")
cols.append("length")
cols.append("sup")
```

```python
df2=df2[cols]
```
